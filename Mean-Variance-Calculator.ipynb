## **1. Get User Input**

```python
# Ask the user to enter numbers separated by commas
user_input = input("Enter numbers separated by commas: ")

# Split the input string into a list using comma as separator
numbers_str = user_input.split(",")

# Convert each value from string to float (so we can calculate)
numbers = [float(num) for num in numbers_str]
```

## **2. Calculate the Mean**

```python
# Calculate the total sum of all numbers
total = sum(numbers)

# Count how many numbers were entered
count = len(numbers)

# Calculate the mean (average)
mean = total / count
```

## **3. Calculate the Variance**

```python
# Create a list of squared differences from the mean
squared_differences = [(num - mean) ** 2 for num in numbers]

# Calculate the variance (average of squared differences)
variance = sum(squared_differences) / count
```

## **4. Display Results**

```python
# Print the results in a clear format
print("Mean:", mean)
print("Variance:", variance)
```
